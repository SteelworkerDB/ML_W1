```python
#Carga de librerías

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.impute import SimpleImputer
```

```python
# Carga de datos
filename='/content/abalone.data'
header=['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight',\
        'Viscera weight','Shell weight','Rings']
df=pd.read_csv(filename, names=header)
df.head(10)
```

**1. Separar los datos en un matriz de características (X) y un vector objetivo (y).**

```python
#La columna a predecir es la de Anillos, el resto será la matriz de
#características

X=df.drop(columns='Rings')
y=df['Rings']

```

**2. Crear un train test split en los datos. Utilicen un número aleatorio 42 por coherencia.**

```python
#Preparación
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

**3. Utilizar transformadores de columna para transformar las columnas adecuadas**

```python
#Selector de columnas tipo object
cat_selector = make_column_selector(dtype_include='object')

#Selector de columnas tipo numérico
num_selector = make_column_selector(dtype_include='number')
```

```python
# Instanciando el escalador estándar y el codificador one hot
scaler = StandardScaler()
ohe = OneHotEncoder(handle_unknown='ignore')
```

```python
#tuplas de los transformadores
num_tuple = (scaler, num_selector)
cat_tuple = (ohe, cat_selector)
```

```python
#Preparar instancia del transformador de columna
col_transformer = make_column_transformer(num_tuple, cat_tuple, remainder = 'passthrough')
```

```python
#Ajuste del transformador a la matriz de entrenamiento
col_transformer.fit(X_train)
```

```python
#Aplicar el transformador de columnas a las dos matrices
X_train_processed = col_transformer.transform(X_train)
X_test_processed = col_transformer.transform(X_test)
X_train_processed[:5] #5 primeras filas
```

```python
#Transformación a dataframe para verificar los datos

X_train_df = pd.DataFrame(X_train_processed)
X_train_df.head()
```
