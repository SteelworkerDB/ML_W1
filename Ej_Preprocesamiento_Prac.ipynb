```python
#Carga de librerías

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
# Carga de datos
filename='/content/insurance.csv'
df=pd.read_csv(filename)
df.head()
```

**1.**

```python
#La columna a predecir es la de charges, el resto será la matriz de
#características

X=df.drop(columns='charges')
y=df['charges']

```

**2.**

```python
#Preparación
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=31)
```

**3.**

  Identificación de característica:

  Age: Numérica

  Sex: Nominal

  BMI: Numérica

  children: Numérica
  
  Smoker: Nominal
  
  Region: Nominal

**4.** Codificación one-hot cualquier característica nominal

```python
#Filtro de columnas tipo object
cat_selector = make_column_selector(dtype_include='object')

#Se filtran las columnas de X train en la matriz train cat data
train_cat_data = X_train[cat_selector(X_train)]

#Se filtran las columnas en Xtest en la matriz test cat data
test_cat_data= X_test[cat_selector(X_test)]
train_cat_data
```

```python
# Instancia de OneHotEncoder
ohe_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

#Ajuste dle encoder a los datos de entrenamiento
ohe_encoder.fit(train_cat_data)

#Aplico encoder
train_ohe=ohe_encoder.transform(train_cat_data)
test_ohe=ohe_encoder.transform(test_cat_data)
train_ohe
```

```python
# Para devolver a un dataframe

#Obtener los nombres de las nuevas columnas
ohe_column_names=ohe_encoder.get_feature_names_out(train_cat_data.columns)

#transformar a dataframes
train_ohe = pd.DataFrame(train_ohe, columns=ohe_column_names)
test_ohe = pd.DataFrame(test_ohe, columns=ohe_column_names)
train_ohe
```

**5.** Estandarización de datos numéricos

```python
#Tomamos los datos numéricos
#Filtro de columnas tipo numerico
num_selector = make_column_selector(dtype_include='number')


#Se filtran las columnas de X train en la matriz train nums
train_nums = X_train[num_selector(X_train)].reset_index(drop=True)

#Se filtran las columnas en Xtest en la matriz test cat data
test_nums= X_test[num_selector(X_test)].reset_index(drop=True)

#Escalar/Estandarizar
#Instancia
scaler = StandardScaler()

#Ajuste del escalador
scaler.fit(train_nums)

#Aplicación del scaler
train_scaled = scaler.transform(train_nums)

test_scaled = scaler.transform(test_nums)

train_scaled[:5]
```

```python
#Pasando los datos numéricos estandarizados a dataframe

Train_nums_sc=pd.DataFrame(train_scaled,columns=train_nums.columns)
Test_nums_sc=pd.DataFrame(test_scaled,columns=test_nums.columns)
Train_nums_sc.head()
```
